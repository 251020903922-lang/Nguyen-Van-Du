```python
#2.Thu·∫≠t to√°n trung b√¨nh c·ªông c·ªßa n s·ªë

n = int(input("Nh·∫≠p s·ªë l∆∞·ª£ng n: "))
tong = 0
for i in range(n):
    x = float(input(f"Nh·∫≠p s·ªë th·ª© {i + 1}: "))
    tong += x
tbc = tong / n
print("Trung b√¨nh c·ªông =",tbc)
```

```python
#3.Vi·∫øt thu·∫≠t to√°n ƒë·∫øm s·ªë trong danh s√°ch

n = list(map(int,input("Nh·∫≠p d√£y s·ªë : ").split()))
so_le = 0
so_chan = 0
for i in n:
  if i % 2 == 0:
    so_chan += 1
  else:
    so_le += 1
print("C√≥",so_le,("s·ªë l·∫ª"))
print("C√≥",so_chan,"s·ªë ch·∫µn")
```

```python
#4.Vi·∫øt thu·∫≠t to√°n t√¨m gi√° tr·ªã l·ªõn nh·∫•t v√† nh·ªè nh·∫•t

n = list(map(int,input("Nh·∫≠p d√£y s·ªë : ").split()))
print("Gi√° tr·ªã l·ªõn nh·∫•t l√† :",max(n))
print("Gi√° tr·ªã nh·ªè nh·∫•t l√† :", min(n))
```

```python
#5.Vi·∫øt thu·∫≠t to√°n s·∫Øp x·∫øp tƒÉng d·∫ßn 5 s·ªë

n = list(map(int, input("Nh·∫≠p c√°c s·ªë t·ª± nhi√™n: ").split()))
n.sort()
print("S·∫Øp x·∫øp s·ªë tƒÉng d·∫ßn:", n)
```

```python
#6.Vi·∫øt thu·∫≠t to√°n t√≠nh giai th·ª´a b·∫±ng ƒë·ªá quy

def giaithua(n):
    if n == 0:
        return 1
    else:
        return n * giaithua(n - 1)
n = int(input("Nh·∫≠p s·ªë nguy√™n d∆∞∆°ng n : "))
if n < 0:
    print("Vui l√≤ng nh·∫≠p s·ªë nguy√™n d∆∞∆°ng")
else:
    print(f"{n}! =", giaithua(n))
```
