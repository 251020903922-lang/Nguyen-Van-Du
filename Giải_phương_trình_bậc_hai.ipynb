```python
import math

print("Giải phương trình bậc hai ax^2 + bx + c = 0")

a = float(input("a="))
b = float(input("b="))
c = float(input("c="))

if a == 0:
    print("Đây không phải phương trình bậc hai (a = 0).")
else:
    delta = b**2 - 4*a*c

    if delta > 0:
      print("Phương trình có nghiệm phên biệt")
      x1=(-b+math.sqrt(delta))/2*a
      x2=(-b-math.sqrt(delta))/2*a
      print("x1=",x1)
      print("x2=",x2)
    if delta < 0:
        print("Phương trình vô nghiệm (delta < 0).")
    elif delta == 0:
        x = -b / (2*a)
        print("Phương trình có nghiệm kép x =", x)

```
