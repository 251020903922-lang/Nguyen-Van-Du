```python
#BT1. Kiểm tra học lực sinh viên

GPA = float(input("Nhập điểm sinh viên : "))
if GPA > 10 or GPA < 0:
  print("Điểm không hợp lệ")
elif GPA < 4:
  print("Học lực yếu")
elif GPA < 5.5:
  print("Học lực trung bình")
elif GPA < 7.0:
  print("Học lực khá")
elif GPA < 8.5:
  print("Học lực giỏi")
else:
  print("Học lực xuất sắc")
```

```python
#BT2. Tính tổng các số chẵn trong khoảng (for)

n = int(input("Nhập số nguyên dương n = "))
tong = 0
for i in range (1, n+1):
  if i % 2 == 0:
    tong += i
print("Tổng các số từ 1 đến",n,"là",tong)
```

```python
#BT3. Đếm số lần nhập đúng mật khẩu (while)

mat_khau_dung = "python123"
dem = 0
while dem < 3:
    mat_khau = input("Nhập mật khẩu: ")
    dem += 1
    if mat_khau == mat_khau_dung:
        print("Đăng nhập thành công")
        break
    else:
        print("Sai mật khẩu")
if dem == 3 and mat_khau != mat_khau_dung:
    print("Tài khoản bị khóa")
```
