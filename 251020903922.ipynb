```python
#Bài1

n = int(input("Nhập số nguyên dương n = "))
tong = 0
if n <= 0:
    print("Số không hợp lệ")
else:
    for i in range(2, n + 1, 2):
        tong += i
    print(f"Tổng các số chẵn nhỏ hơn hoặc bằng {n} là: {tong}")
```

```python
#Bài2

s = input("Nhập chuỗi: ")
words = s.split()
so_tu = len(words)
tu_dai_nhat = words[0]
for tu in words:
    if len(tu) > len(tu_dai_nhat):
        tu_dai_nhat = tu
print("Words:", so_tu)
print("Longest word:", tu_dai_nhat)
```

```python
#Bài3

arr = list(map(int, input("Nhập các số: ").split()))
arr_khong_trung = list(set(arr))
arr_khong_trung.sort()
if len(arr_khong_trung) < 2:
    print("Không đủ phần tử để tìm số lớn thứ hai")
else:
    so_lon_thu_hai = arr_khong_trung[-2]
    print("Sorted list:", arr_khong_trung)
    print("Second largest:", so_lon_thu_hai)
```

```python
#Bài4

danh_sach_diem = list(map(float,input("Nhập danh sách điểm : ").split()))
diem_hop_le = [ i for i in danh_sach_diem if 0 <= i <= 10 ]
if len(diem_hop_le) == 0:
  print("Không có điểm hợp lệ")
else:
  diem_trung_binh =sum(diem_hop_le) / len(diem_hop_le)
  dat = sum(1 for i in diem_hop_le if i >= 5 )
print("Danh sách điểm hợp lệ:", diem_hop_le)
print("Điểm trung bình:", round(diem_trung_binh, 2))
print("Số sinh viên đạt (≥ 5):", dat)
```
